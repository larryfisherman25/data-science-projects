# Activity Recognition from Single Chest-Mounted Accelerometer Data Set

### **Fonte:**

Com um acelerômetro não calibrado, os dados são recolhidos de 15 participantes que realizam 7 actividades. O conjunto de dados fornece desafios para a identificação e autenticação de pessoas que utilizam padrões de movimento.

### **Informação sobre o conjunto de dados:**


* O conjunto de dados recolhe dados a partir de um acelerómetro utilizável montado no peito
* Frequência de amostragem do acelerómetro: 52 Hz
* Os dados do acelerômetro não são calibrados
* Número de Participantes: 15
* Número de Actividades: 7
* Formato dos dados: CSV


### **Informação de Atributos:**

#### Os dados são separados por participante, cada ficheiro contém as seguintes informações:
* número sequencial, x aceleração, y aceleração, z aceleração, etiqueta



#### Os rótulos são codificados por números
```
 1: Trabalhar no computador
 2: Levantar, Caminhar e Subir e Descer escadas
 3: Em pé
 4: Andar a pé
 5: Subir e descer escadas
 6: Caminhar e Falar com Alguém
 7: Conversar em pé
```

## 2.0 - Preparação dos dados

Trata-se de 15 usuários (15 datasets) para análisar, colocaremos todos no mesmo dataframe e diferenciaremos os mesmos pelo seu id (```User_id```)

````python
# aglutina todos os cvs no mesmo caminho
files_path = glob(r"./datasets/*.csv")
print(files_path)
````

````python
def gerar_df(files_path:list) -> pd.DataFrame:
    """
    função que inclui user_id com base no número do arquivo csv
    """
    df_users = pd.DataFrame()
    
    for index, path in enumerate(files_path):
        df = pd.read_csv(path, header=None)
        df['user_id'] = int(re.sub(r'[^0-9]', "", path))
        df_users = pd.concat([df_users, df.iloc[:, 1:]])
        
    return df_users
````

````python
data = gerar_df(files_path)

data.columns = ['x_acceleration',
                'y_acceleration',
                'z_acceleration',
                'label',
                'user_id',
               ]

data.index = range(len(data))
````

````python
data = data.reindex(columns=["user_id",
                             "x_acceleration",
                             "y_acceleration",
                             "z_acceleration",
                             "label",
                            ])
````

## 3.0 - Engenharia de Parâmetros

````python
mag_x = np.square(data['x_acceleration'])
mag_y = np.square(data['y_acceleration'])
mag_z = np.square(data['z_acceleration'])

mag_total = np.sqrt(mag_x + mag_y + mag_z)

# criando nova coluna
data['mag_acceleration'] = mag_total

# organizando dataframe
data = data.reindex(['user_id',
                     'x_acceleration',
                     'y_acceleration',
                     'z_acceleration',
                     'mag_acceleration',
                     'label'
                    ], axis=1)
````

##### Explorando as variações de aceleração no eixo `x`, `y`, `z`, e `mag` com as mudanças de atividades (`label`)

````python
def plot_user_axis_acc(user):
    for i in range(user.shape[1]):
        plt.figure(figsize=(14, 5))
        plt.subplot(user.shape[1], 1, i+1)
        plt.plot(user[:, i], color='blue')
        plt.show()
    return None

for i in sorted(list(data.user_id.unique())):
    print('user_id', i)
    print('A aceleração nos eixos x,y,z e mag com a mudança de labels')
    plot_user_axis_acc(data.loc[data.user_id == i].iloc[:, 1:].values)
    print(f'Fim do {i} plot\n')
````

* **INSIGHT:** <br>
Todas atividades(`label`s) dependem dos três eixos de aceleração: `x`, `y`, `z`.
